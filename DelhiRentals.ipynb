```python
def remove_outliers_sqft(df):
  df_output=pd.DataFrame()
  for key,subdf in df.groupby('localityName'):
    m=np.mean(subdf.rent_per_sqft)
    st=np.std(subdf.rent_per_sqft)
    gen_df=subdf[(subdf.rent_per_sqft>(m-st)) & (subdf.rent_per_sqft<=(m+st))]
    df_output=pd.concat([df_output,gen_df],ignore_index=True)
  return df_output
data=remove_outliers_sqft(data)
data.describe()
```

```python
data.drop('rent_per_bedroom',axis=1,inplace=True)
```

```python
def remove_bedroom_outliers(df):
  exclude_indices=np.array([])
  for location,location_df in df.groupby('localityName'):
    bedstats={}
    for bed,bed_df in location_df.groupby('bedrooms'):
      bedstats={
          'mean':np.mean(bed_df.rent_per_sqft),
          'std':np.std(bed_df.rent_per_sqft),
          'count':bed_df.shape[0]
      }
    for bed,bed_df in location_df.groupby('bedrooms'):
      stats=bedstats.get(bed-1)
      if stats and stats['count']>5:
        exclude_indices=np.append(exclude_indices,bed_df[bed_df.rent_per_sqft<(stats['mean'])].index.values)
  return df.drop(exclude_indices,axis='index')
```

```python
data=remove_bedroom_outliers(data)
```

```python
X=data.drop(columns=['rent'])
y=data['rent']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print(X_train)
print(X_test)
```

```python
#applying Linear Regression
```

```python
column_trans=make_column_transformer((OneHotEncoder(sparse=False),['localityName']),remainder='passthrough')
```

```python
scaler=StandardScaler()
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lr=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_lr)
```

```python
#apply lasso
```

```python
lasso=Lasso()
```

```python
pipe=make_pipeline(column_trans,scaler,lasso)
```

```python
pipe.fit(X_train,y_train)
y_pred_lasso=pipe.predict(X_test)
r2_score(y_test,y_pred_lasso)
```

```python
r2_score(y_test,y_pred_lasso)
```

```python
#apply ridge
```

```python
ridge=Ridge()
```

```python
pipe=make_pipeline(column_trans,scaler,ridge)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_ridge=pipe.predict(X_test)
r2_score(y_test,y_pred_ridge)
```
